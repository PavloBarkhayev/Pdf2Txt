```python
# https://ru.wikipedia.org/wiki/%D0%A2%D1%80%D0%B0%D0%BD%D1%81%D0%BF%D0%BE%D1%80%D1%82%D0%BD%D0%B0%D1%8F_%D0%B7%D0%B0%D0%B4%D0%B0%D1%87%D0%B0

import heapq
from collections import namedtuple

class Edge:
    def __init__(self, to, rev, cap, cost):
        self.to = to
        self.rev = rev
        self.cap = cap
        self.cost = cost

def minimum_transportation_price(suppliers, consumers, costs):
    n = len(suppliers)
    m = len(consumers)
    
    total_nodes = n + m + 2
    source = total_nodes - 2
    sink = total_nodes - 1
    
    INF = float('inf')

    graph = [[] for _ in range(total_nodes)]

    def add_edge(frm, to, cap, cost):
        graph[frm].append(Edge(to, len(graph[to]), cap, cost))
        graph[to].append(Edge(frm, len(graph[frm]) - 1, 0, -cost))

    # Add edges from source to suppliers
    for i, supply in enumerate(suppliers):
        add_edge(source, i, supply, 0)

    # Add edges from consumers to sink
    for j, demand in enumerate(consumers):
        add_edge(n + j, sink, demand, 0)

    # Add edges from suppliers to consumers with respective costs
    for i in range(n):
        for j in range(m):
            add_edge(i, n + j, float('inf'), costs[i][j])

    # Min-cost max-flow using successive shortest augmenting path with potentials
    def min_cost_flow(s, t, maxf):
        res = 0
        h = [0] * total_nodes  # potential
        prevv = [0] * total_nodes
        preve = [0] * total_nodes

        flow = 0
#        counter = 0
        while flow < maxf:
#            counter += 1
#            if counter > 500:
#                print('500')
#                return None
            dist = [INF] * total_nodes
            dist[s] = 0
            pq = [(0, s)]
            while pq:
                d, v = heapq.heappop(pq)
                if dist[v] < d:
                    continue
                for i, e in enumerate(graph[v]):
                    if e.cap > 0 and dist[e.to] > dist[v] + e.cost + h[v] - h[e.to]:
                        dist[e.to] = dist[v] + e.cost + h[v] - h[e.to]
                        prevv[e.to] = v
                        preve[e.to] = i
                        heapq.heappush(pq, (dist[e.to], e.to))

            if dist[t] == INF:
                return -1

            for v in range(total_nodes):
                h[v] += dist[v]

            d = maxf - flow
            v = t
            while v != s:
                d = min(d, graph[prevv[v]][preve[v]].cap)
                v = prevv[v]
            flow += d
            res += d * h[t]
            v = t
            while v != s:
                e = graph[prevv[v]][preve[v]]
                e.cap -= d
                graph[v][e.rev].cap += d
                v = prevv[v]
        
#        print(f'Counter: {counter}')
        return res

    return min_cost_flow(source, sink, sum(suppliers))
```

```python
import heapq

from collections import deque

class Edge:
    def __init__(self, u, v, cost, capacity):
        self.u = u
        self.v = v
        self.cost = cost
        self.capacity = capacity
        self.flow = 0
        self.rev = None

class NetworkSimplex:
    def __init__(self, supply, demand, cost_matrix):
        self.n = len(supply) + len(demand)
        self.orig_n = self.n
        self.supply = supply
        self.demand = demand
        self.cost_matrix = cost_matrix
        self.adj = [[] for _ in range(self.n + 2)]
        self._build_graph()

    def _add_edge(self, u, v, cost, capacity):
        forward = Edge(u, v, cost, capacity)
        backward = Edge(v, u, -cost, 0)
        forward.rev = backward
        backward.rev = forward
        self.adj[u].append(forward)
        self.adj[v].append(backward)

    def _build_graph(self):
        num_sources = len(self.supply)
        num_sinks = len(self.demand)
        for i in range(num_sources):
            for j in range(num_sinks):
                self._add_edge(i, num_sources + j, self.cost_matrix[i][j], float('inf'))
        super_source = self.n
        super_sink = self.n + 1
        for i in range(num_sources):
            self._add_edge(super_source, i, 0, self.supply[i])
        for j in range(num_sinks):
            self._add_edge(num_sources + j, super_sink, 0, self.demand[j])
        self.n += 2

    def solve(self):
        return self._min_cost_flow(self.n - 2, self.n - 1)

    def _min_cost_flow(self, source, sink):
        flow = 0
        cost = 0
        while True:
            dist = [float('inf')] * self.n
            inqueue = [False] * self.n
            prev = [None] * self.n
            dist[source] = 0
            q = deque([source])

            while q:
                u = q.popleft()
                inqueue[u] = False
                for e in self.adj[u]:
                    if e.flow < e.capacity and dist[e.v] > dist[u] + e.cost:
                        dist[e.v] = dist[u] + e.cost
                        prev[e.v] = e
                        if not inqueue[e.v]:
                            q.append(e.v)
                            inqueue[e.v] = True

            if dist[sink] == float('inf'):
                break

            # Find minimum residual capacity on the path
            aug_flow = float('inf')
            v = sink
            while v != source:
                e = prev[v]
                aug_flow = min(aug_flow, e.capacity - e.flow)
                v = e.u

            # Augment the flow along the path
            v = sink
            while v != source:
                e = prev[v]
                e.flow += aug_flow
                e.rev.flow -= aug_flow
                cost += e.cost * aug_flow
                v = e.u

            flow += aug_flow

        return cost


def minimum_transportation_price(suppliers, consumers, costs):
    return NetworkSimplex(suppliers, consumers, costs).solve()
```

```python
from collections import deque

def solve_puzzle(board,pieces):
    n = len(board)
    grid = [[1 if c == ' ' else 0 for c in row] for row in board]
    filled = [[False]*n for _ in range(n)]
    
    # Preprocess and group pieces by size
    piece_variants = []
    for idx, (h, w) in enumerate(pieces):
        pset = set()
        pset.add((h, w, 0))
        if h != w:
            pset.add((w, h, 1))
        piece_variants.append((idx, list(pset)))

    # Sort by area descending
    piece_variants.sort(key=lambda x: -max(w*h for w, h, _ in x[1]))

    result = [None] * len(pieces)

    # Bitmask helpers for board
    def fits(x: int, y: int, h: int, w: int) -> bool:
        if x + h > n or y + w > n:
            return False
        for dx in range(h):
            for dy in range(w):
                if grid[x+dx][y+dy] == 1 or filled[x+dx][y+dy]:
                    return False
        return True

    def place(x: int, y: int, h: int, w: int, val: bool):
        for dx in range(h):
            for dy in range(w):
                filled[x+dx][y+dy] = val

    def backtrack(i: int) -> bool:
        if i == len(piece_variants):
            return True

        pid, orientations = piece_variants[i]
        for h, w, rot in orientations:
            for x in range(n - h + 1):
                for y in range(n - w + 1):
                    if fits(x, y, h, w):
                        place(x, y, h, w, True)
                        result[pid] = [y, x, rot]
                        if backtrack(i + 1):
                            return True
                        place(x, y, h, w, False)
        return False

    backtrack(0)
    return [[b,a,c] for a,b,c in result]
```
